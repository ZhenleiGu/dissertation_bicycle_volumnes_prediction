```python
site_df = pd.read_excel('./drive/MyDrive/Streetview/Count sites list.xlsx')
site_df.head()
```

```python
result = pd.merge(g_df_sum, site_df, how='left', on=["Site ID"])
result.head()
```

```python
result_cor = result[['Site ID','Location','Total cycles','Easting','Northing']]
#result_cor = result[['Site ID','Location','Number of private cycles','Number of cycle hire bikes','Total cycles','Easting','Northing']]
result_cor[['lat','lng']]=0
result_cor.head()
```

```python
from pyproj import Transformer, Proj, transform

def cor_transfer(e, n):
  WGS84 = Proj(init='EPSG:4326')
  p = Proj(init="EPSG:27700")
  # utm
  lon, lat = transform(p, WGS84, e, n)
  # print("x:", x, "y:", y)
  # print("lat:", lat, "lon:", lon)
  return lat, lon

print(cor_transfer(result_cor['Easting'][1],result_cor['Northing'][1]))
```

```python
for index, row in result_cor.iterrows():
      #print(list(df.columns)[4])
      print(row[0])
      lat,lng = cor_transfer(row['Easting'],row['Northing'])
      result_cor.loc[index, 'lat'] = lat
      result_cor.loc[index, 'lng'] = lng
```
