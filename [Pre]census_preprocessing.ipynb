```python
import pandas as pd
import numpy as np
import time

pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('./drive/MyDrive/Streetview/lsoa-data.csv')
df.head()
```

```python
df['Road Casualties;2010-2014;Total']=df.iloc[:,-20:].filter(regex='Total').sum(axis=1)
df['Road Casualties;2010-2014;Fatal']=df.iloc[:,-20:].filter(regex='Fatal').sum(axis=1)
df['Road Casualties;2010-2014;Serious']=df.iloc[:,-20:].filter(regex='Serious').sum(axis=1)
df['Road Casualties;2010-2014;Slight']=df.iloc[:,-20:].filter(regex='Slight').sum(axis=1)
```

```python
df.drop(columns=df.columns[-24:-4], 
        axis=1, 
        inplace=True)
df.iloc[:,-10:].head()
```

```python
df.to_csv('./drive/MyDrive/Streetview/lsoa-data-new.csv', index=False)
```
