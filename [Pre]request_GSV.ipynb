**This file requets the street view of each counting location via the Google GSV API**

```python
import requests
import os
import pandas as pd
```

```python
from google.colab import drive 
drive.mount('/content/drive/')
```

```python
meta_base = 'https://maps.googleapis.com/maps/api/streetview/metadata?'
pic_base = 'https://maps.googleapis.com/maps/api/streetview?'
api_key = ''
# change to your api key
```

```python
df = pd.read_csv('./drive/MyDrive/Streetview/outer_2019_id_final.csv')
df.head()
```

```python
for index, row in df.iterrows():
  #print(row['LSOA11NM'].split()[0])
  print(row['Site ID'])
  for i in ['0','90','180','270']:
    pic_params = {'key': api_key,
                  'location': row['Location']+','+row['LSOA11NM'].split()[0],
                  'size': "640x640",
                  'heading':i,
                  'fov':"90"}
    pic_response = requests.get(pic_base, params=pic_params)
    save_path = './drive/MyDrive/Streetview/Outer_GSV_data/' + row['Site ID'] + '_' + i + '.jpg'
    os.makedirs(os.path.dirname(save_path), exist_ok=True)
    with open(save_path, 'wb') as file:
        file.write(pic_response.content)
  
```

```python
for j in list(df['Location']):
  for i in ['0','90','180','270']:
    pic_params = {'key': api_key,
                  'location': j+',London',
                  'size': "640x640",
                  'heading':i,
                  'fov':"90"}
    pic_response = requests.get(pic_base, params=pic_params)
    save_path = './drive/MyDrive/Streetview/GSV_data/' + j + '_' + i + '.jpg'
    os.makedirs(os.path.dirname(save_path), exist_ok=True)
    with open(save_path, 'wb') as file:
        file.write(pic_response.content)
```
