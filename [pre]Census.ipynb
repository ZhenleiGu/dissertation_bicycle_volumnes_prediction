```python
df = pd.read_csv('./drive/MyDrive/Streetview/inner_2019_id_avg_cor.csv')
df.head()
```

```python
gdf = geopandas.GeoDataFrame(df, geometry=geopandas.points_from_xy(df['lng'], df['lat']), crs="EPSG:4326")
gdf.head()
```

```python
# gdf = gdf[gdf['lat']>51.4]
# gdf = gdf[gdf['lat']<51.6]
# gdf = gdf[gdf['lng']>-0.3]
# gdf = gdf[gdf['lng']<0.02]
```

```python
plt.figure(figsize=[20,20]) 
gdf.plot(color='red')
plt.show()
```

```python
ax = lsoa.plot(color='white', edgecolor='black')

gdf.plot(ax=ax, color='red')

plt.show()
```

```python
gdf['LSOA11CD'] = 0
gdf['LSOA11NM'] = 0
gdf.head()
```

```python
for index, row in gdf.iterrows():
  #print(row['geometry'])
  for index2, row2 in lsoa.iterrows():
    if row2['geometry'].contains(row['geometry']):
      print(row2['LSOA11CD'],row2['LSOA11NM'])
      gdf.loc[index, 'LSOA11CD'] = row2['LSOA11CD']
      gdf.loc[index, 'LSOA11NM'] = row2['LSOA11NM']
```
